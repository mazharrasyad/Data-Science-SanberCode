14. Pilih lah kode-kode untuk melengkapi fungsi berikut, berurutan sesuai fungsi alur jalannya fungsi tersebut : *

image.png

```
luas_tanah = {
    'tanah_a' : 20,
    'tanah_b' : 30,
    'tanah_c' : 40
}
def print_key_val(dict_obj):
  for key, value in dict_obj.items():
    print('key : ' + key + ' , Val : ' + str(value))
  
print_key_val(luas_tanah)
```
